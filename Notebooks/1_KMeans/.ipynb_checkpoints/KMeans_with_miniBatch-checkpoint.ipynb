# miniBtach mit PCA

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD


data = df['content']
tf_idf_vectorizor = TfidfVectorizer(max_features=1000,
                                    stop_words='english')


tf_idf = tf_idf_vectorizor.fit_transform(data)
tf_idf_norm = normalize(tf_idf)
tf_idf_array = tf_idf_norm.toarray()
# sklearn_pca = PCA(n_components = 5)
# data= sklearn_pca.fit_transform(tf_idf_array)
```

```python
from sklearn.cluster import MiniBatchKMeans

sklearn_pca = PCA(n_components = 5)
Y_sklearn = sklearn_pca.fit_transform(tf_idf_array)
kmeans = MiniBatchKMeans(n_clusters=5, random_state=0, batch_size=6)
fitted = kmeans.fit(Y_sklearn)
prediction = kmeans.predict(Y_sklearn)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(Y_sklearn[:, 0], Y_sklearn[:, 1], c=prediction, s=70, cmap='viridis')
plt.title('PCA')
#plt.savefig('K_Means/PCA_Mini-Batch_Lyrics.png', bbox_inches = "tight")
```

# miniBatch mit TSNE

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.manifold import TSNE

from sklearn.cluster import MiniBatchKMeans
```

```python
data = TfidfVectorizer(max_features=1000,
                       stop_words='english').fit_transform(df.content)
```

```python
X_reduced = TruncatedSVD(n_components=7, random_state=0).fit_transform(data)
```

```python
perplexity = [50]
for p in perplexity:
    X_embedded = TSNE(n_components=2, perplexity=p, n_iter=1000,learning_rate=10, verbose=2).fit_transform(X_reduced)
```

```python
kmeans = MiniBatchKMeans(n_clusters=5, max_iter=600, random_state=0, batch_size=6)
fitted = kmeans.fit(X_embedded)
prediction = kmeans.predict(X_embedded)
```

```python
markers = ('o', 'v', '^', 's', 'p', '*','d', 'P', 'X')
plt.figure(figsize=(15,10))
sns.scatterplot(X_embedded[:, 0], X_embedded[:, 1],hue=prediction, style=df.category, markers=markers, palette=sns.color_palette("hls", len(np.unique(prediction))))

plt.title('TSNE')
#plt.savefig('../../Bilder/KMeans/kmeans_tsne_categorymarker_stemming_5_Cluster.png', bbox_inches = "tight")
```

# miniBatch mit UMAP

```python
from sklearn.feature_extraction.text import TfidfVectorizer

from yellowbrick.datasets import load_hobbies
from yellowbrick.text import UMAPVisualizer
import umap
import sklearn.cluster as cluster
```

```python
def tfidf(data):
    tf_idf_vectorizor = TfidfVectorizer(max_features=1000,
                                        stop_words='english')
    tf_idf = tf_idf_vectorizor.fit_transform(data)
    tf_idf_norm = normalize(tf_idf)
    tf_idf_array = tf_idf_norm.toarray()
    return tf_idf_array
```

```python
docs = tfidf(df.content)
labels = df.category
```

```python
standard_embedding = umap.UMAP(random_state=42).fit_transform(docs)
```

```python
kmeans = MiniBatchKMeans(n_clusters=5, max_iter=600, random_state=0, batch_size=6)
fitted = kmeans.fit(standard_embedding)
prediction = kmeans.predict(standard_embedding)

#kmeans_labels = kmeans(n_clusters=5, max_iter=600, algorithm='auto').fit_predict(standard_embedding)
```

```python
markers = ('o', 'v', '^', 's', 'p', '*','d', 'P', 'X')
plt.figure(figsize=(15,10))
sns.scatterplot(standard_embedding[:, 0], standard_embedding[:, 1],hue=prediction, style=df.category, 
                markers=markers, palette=sns.color_palette("hls", len(np.unique(prediction))))

plt.title('MiniBatch with UMAP')
#plt.savefig('K_Means/minibatch_tsne_pos_ohne_genre.png')
```
